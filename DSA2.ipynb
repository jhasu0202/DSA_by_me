#Functions

```python
def greet(name, greeting='Hello'):
 return f'{greeting}, {name}!'
name=input("enter name")
greet(name)
```

```python
def min_max(arr):
 return min(arr), max(arr) # returns a tuple
min_max([3,1,4,1,5])
```

```python
def total(*nums):
  return sum(nums)
total(1, 2, 3, 4)
```

```python
def profile(**info):
  for k, v in info.items():
    print(f'{k}: {v}')
profile(name='Alice', age=22)
```

```python
#lambda function
evens = list(filter(lambda x: x%2==0, range(10)))
print(evens)
```

```python
square = lambda x: x * x
print(square(10))
```

```python
# Nested function + closure
def multiplier(factor):
  def multiply(n):
    return n * factor
  return multiply
triple = multiplier(3)
print(triple(7))
```

RECURSION

```python
# Factorial
def factorial(n):
 if n == 0: return 1 # base case
 return n * factorial(n-1) # recursive case
factorial(5)
```

```python
# Fibonacci (naive O(2^n))
def fib(n):
 if n <= 1: return n
 return fib(n-1) + fib(n-2)
factorial(5)
```

```python
# Sum of list recursively
def rsum(arr):
 if not arr: return 0
 return arr[0] + rsum(arr[1:])
rsum([1,2,3,4])
```

```python
# Power: x^n recursively
def power(x, n):
 if n == 0: return 1
 half = power(x, n//2)
 if n % 2 == 0: return half * half
 else: return x * half * half # O(log n)
power(2,3)
```

```python
# Reverse string recursively
def rev(s):
 if len(s) <= 1: return s
 return rev(s[1:]) + s[0]
rev("jhasu")
```
