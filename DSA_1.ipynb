Asymptotic notations give us a mathematical language to describe algorithm efficiency without needing to
run the code. Three primary notations:
Big-O (O) — upper bound: worst case
Omega (Ω) — lower bound: best case
Theta (Θ) — tight bound: average case

 Linear Search: O(n) | Binary Search: O(log n)

Big-O Notation
f(n) = O(g(n)) means there exist constants c > 0 and n0 such that f(n) ≤ c·g(n) for all n ≥ n0. Big-O captures
the worst-case upper bound

Omega (Ω) Notation
f(n) = Ω(g(n)) means f(n) grows at least as fast as g(n). It gives the best-case (lower bound) guarantee.


Theta (Θ) Notation
f(n) = Θ(g(n)) when f(n) is both O(g(n)) and Ω(g(n)). The algorithm always runs in exactly that order of growth, regardless of input.


For recursive algorithms we write a recurrence relation and solve it. Common method: Master Theorem
and Recursion Tree.

 Factorial — T(n) = T(n-1) + O(1) → O(n)

 Fibonacci (naive) — T(n) = T(n-1)+T(n-2)+O(1) → O(2^n)


 Binary search recursive — T(n) = T(n/2)+O(1) → O(log n)


Merge Sort — T(n) = 2T(n/2)+O(n) → O(n log n

 Master Theorem: T(n)=aT(n/b)+f(n).
 If f(n)=O(n^log_b(a)) → T(n)=Θ(n^log_b(a) · log n).
Merge sort: a=2,b=2
→ log_2(2)=1 → Θ(n log n).


Space Complexity
Space complexity measures extra memory an algorithm uses (excluding input). Includes stack frames,
auxiliary arrays, and variables

O(1) space — only fixed variables


O(n) space — new list of size n

 O(n) stack space — n recursive calls on stack


 O(log n) stack space — binary search recursive

Given weight(kg) and height(m), compute BMI=weight/height^2. Print: Underweight(<18.5),
Normal(18.5-24.9), Overweight(25-29.9), Obese(>=30).

```python
w = 30
h = 20
bmi = w/(h**2)
if bmi<18.5:
  print("Underweight")
elif (bmi>=18.5 & bmi<24.9):
  print("Normal")
elif (bmi>24 & bmi<=29.9):
  print("Overweight")
else:
  print("Obese")
```

```python
units = int(input("Enter the number of units consumed: "))
bill = 0

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = (100 * 1.5) + ((units - 100) * 2)
elif units <= 300:
    bill = (100 * 1.5) + (100 * 2) + ((units - 200) * 3)
else:
    bill = (100 * 1.5) + (100 * 2) + (100 * 3) + ((units - 300) * 4)

print(f"Your electricity bill for {units} units is Rs {bill:.2f}")
```

```python
# Number pyramid
for i in range(1, n+1):
  print(' '*(n-i), end='')
  for j in range(1, 2*i):
    print(j if j<=i else 2*i-j, end='')
  print()
```

```python
# Prime sieve
def sieve(n):
  is_prime = [True] * (n + 1)
  is_prime[0] = is_prime[1] = False
  for i in range(2, int(n**0.5) + 1):
    if is_prime[i]:
      for j in range(i * i, n + 1, i):
        is_prime[j] = False
  return [i for i in range(n + 1) if is_prime[i]]

# Example usage:
primes_up_to_30 = sieve(30)
print(f"Prime numbers up to 30: {primes_up_to_30}")
```
