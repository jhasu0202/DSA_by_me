Python's list is a dynamic array: it can grow and shrink at runtime



Python does NOT store the
actual objects inside the array; instead it stores pointers (references) to objects on the heap

When
the internal buffer fills up, Python allocates a new, larger buffer (typically ~1.125× the old size) and
copies all pointers — this is called resizing.
