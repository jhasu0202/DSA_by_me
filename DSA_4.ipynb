1.2 Reverse a Number

```python
def reverse_number(n):
  neg = n<0
  n = abs(n)
  r = 0
  while n>0:
    d=n%10
    r=r*10+d
    n=n//10
  if neg:
    r =-r
  if r < -2**31 or r > 2**31 - 1:
    return 0
  return r
reverse_number(1212)
# Time: O(d) Space: O(1)
```

 Palindrome Number

```python
def palindrome(n):
  m = n
  r = 0
  while n>0:
    d=n%10
    r=r*10+d
    n=n//10
  if r == m :
    return 1
  else :
    return 0
palindrome(1212)
```

```python
def is_palindrome_number(n):
  if n < 0:
    return False # negative numbers can never be palindromes
  original = n
  reversed_num = 0
  while n > 0:
    digit = n % 10
    reversed_num = reversed_num * 10 + digit
    n = n // 10
  return original == reversed_num
is_palindrome_number(1211)
# Time: O(d) Space: O(1)
```

```python
def is_palindrome_half(n):
  if n < 0 or (n % 10 == 0 and n != 0):
    return False # numbers ending in 0 can't be palindromes (except 0 itself)
  reversed_half = 0
  while n > reversed_half:
    reversed_half = reversed_half * 10 + n % 10
    n //= 10
  # even digit count: n == reversed_half
  # odd digit count: middle digit dropped -> n == reversed_half // 10
  return n == reversed_half or n == reversed_half // 10
is_palindrome_half(12321)
# Time: O(d/2) -- still O(d) Space: O(1)
```

GCD (Greatest Common Divisor) of Two Numbers

```python
 #Brute Force
 def gcd(a,b):
   s=min(a,b)
   for i in range(s,0,-1):
    if a%i==0 & b%i==0:
      return i
   return 1
 gcd(15,20)
# Time: O(min(a, b)) Space: O(1)

```

```python
#Euclidean Algorithm (optimal)
#Core idea: gcd(a, b) = gcd(b, a % b) , and gcd(a, 0) = a
def gcd(a, b):
  while b != 0:
    a, b = b, a % b
  return a
# Time: O(log(min(a, b))) Space: O(1)
# Recursive version (same logic, uses the call stack instead of a loop)
def gcd_recursive(a, b):
  if b == 0:
    return a
  return gcd_recursive(b, a % b)
gcd(3,18)
gcd_recursive(22,44)
```

 Check for Prime Numbe

```python
def is_prime_brute(n):
  if n < 2:
    return False
  for i in range(2, n):
    if n % i == 0:
      return False
  return True
# Time: O(n) Space: O(1)
```

```python
def is_prime(n):
  if n < 2:
    return False
  if n in (2, 3):
    return True
  if n % 2 == 0:
    return False
  i = 3
  while i * i <= n:
    if n % i == 0:
      return False
    i += 2 # skip even numbers -- only check odd divisors
  return True
# Time: O(sqrt(n)) Space: O(1)
```

* Sieve of Eratosthenes
* If you need to check primality for every number up to n (not just one number), checking each individually with the sqrt(n)
method is O(n · sqrt(n)) total — too slow. The Sieve of Eratosthenes instead marks all multiples of each prime as non-prime
in one pass, giving O(n log log n) total for all numbers up to n. (This connects to LeetCode 204 — Count Primes, and was also
mentioned briefly in your Week 1 basics PDF, Day 2 — Loops.)

```python
def sieve_of_eratosthenes(n):
  is_prime_arr = [True] * (n + 1)
  is_prime_arr[0] = is_prime_arr[1] = False

  for i in range(2, int(n ** 0.5) + 1):
    if is_prime_arr[i]:
      for multiple in range(i * i, n + 1, i):
        is_prime_arr[multiple] = False

  return [i for i in range(n + 1) if is_prime_arr[i]]
# Time: O(n log log n) Space: O(n)
```

**Chapter 2 — Learn Basic Recursion**

The two non-negotiable rules of recursion:
1. Base case — a condition that stops the recursion without calling itself again. Without this, you get infinite recursion
and a RecursionError: maximum recursion depth exceeded .
2. Recursive case — the function calls itself with an input that is provably closer to the base case (usually n-1, or a
smaller slice of the data).





* Shape - 1               
  * do work BEFORE the recursive call
* What it looks like    
  * print(n);
  * recurse(n-1)                         
* Effect
  *   Work happens on the way DOWN the call stack (e.g. print N to 1)
* Shape - 2
  * Do work AFTER the
recursive call
* What it looks like  
  * recurse(n-1);
  * print(n)
* Effect
  * Work happens on the way BACK UP the call stack, i.e. in reverse order (e.g. print 1 to N)


```python
def print_hello(n):
  if n == 0: # base case: nothing left to print
    return
  print("Hello")
  print_hello(n - 1) # recursive case: one fewer to go
print_hello(3)
# Time: O(n) -- n calls
```

```python
def name(n):
  if n==0:
    return
  print("jhasu")
  name(n-1)
name(4)
```

```python
def O_to_N(n):
  if n==0:
    return
  O_to_N(n-1)
  print(n)
O_to_N(4)
# Time: O(n) Space: O(n)
```

```python
def print_n_to_1(n):
  if n == 0:
    return
  print(n) # print FIRST, on the way down
  print_n_to_1(n - 1) # then recurse
print_n_to_1(5)
# Time: O(n) Space: O(n)
```

```python
def sum_n(n):
  if n == 0:
    return 0 # base case: sum of nothing is 0
  return n + sum_n(n - 1) # n plus the sum of everything before it
sum_n(100)
# Time: O(n) Space: O(n)
```

```python
def sum_n_formula(n):
  return n * (n + 1) // 2
# Time: O(1) Space: O(1)
sum_n_formula(50)
```

```python
def factorial(n):
  if n == 0 or n == 1:
    return 1 # base case
  return n * factorial(n - 1)
factorial(5)
# Time: O(n) Space: O(n)
```

```python
def reverse_array(arr, left=0, right=None):
  if right is None:
    right = len(arr) - 1

  if left >= right: # base case: pointers met or crossed
    return

  arr[left], arr[right] = arr[right], arr[left] # swap
  reverse_array(arr, left + 1, right - 1) # recurse inward
arr = [1,2,3,4,5]
reverse_array(arr)      # Call the function
print(arr)              # Print the reversed array
# Time: O(n) Space: O(n) call stack (vs O(1) for the iterative version)
```

```python
def is_palindrome_recursive(s, left=0, right=None):
  if right is None:
    right = len(s) - 1
  if left >= right: # base case: pointers met/crossed -- checked everything
    return True
  if s[left] != s[right]:
    return False
  return is_palindrome_recursive(s, left + 1, right - 1)
s="dad"
is_palindrome_recursive(s)
# Time: O(n) Space: O(n) call stack
```

```python
#Naive Recursion
def fib(n):
  if n <= 1:
    return n # base cases: F(0)=0, F(1)=1
  return fib(n - 1) + fib(n - 2)
# Time: O(2^n) -- exponential! Space: O(n) call stack depth
fib(5)
```

```python
#Most Optimal — Iterative, O(1) space
def fib_iterative(n):
  if n <= 1:
    return n
  a, b = 0, 1
  for _ in range(2, n + 1):
    a, b = b, a + b
  return b
fib_iterative(6)
# Time: O(n) Space: O(1)
```
