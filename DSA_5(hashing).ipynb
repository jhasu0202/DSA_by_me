Chapter 3 — Basic Hashing



precompute counts into an array (or
dict) once, then answer questions about the data instantly (O(1)) instead of re-scanning the whole array every time.


 "how many times does X appear?" more than once, don't loop through the array again
each time — count everything once into a hash array/dict, then just look up the answer.




*   Array-based hashing
(precompute array)
    * Values are small, known-range integers (like 0-9, or 'a'-'z', or 1 to 1000)
*  Dict-based hashing (hashmap)
    * Values are large, negative, or not a small known range (like
arbitrary integers or strings)



**Problem: Given a small array of numbers (say, values between 0 and 12), count how many times each number appears, using arraybased hashing (the simplest possible version, before we generalise to any array in section 3.3).**

```python
def hash1(arr,max):
  hash_arr=[0]*(max+1)

  for num in arr:
    hash_arr[num]+=1
  return hash_arr
# Example usage:
arr = [1, 2, 2, 5, 3, 3, 5]
hash_arr = hash1(arr, max=5)
print(hash_arr) # [0, 1, 2, 3, 0, 1]
print(hash_arr[3]) # how many times did 3 appear? -> 3, in O(1)
# Time to build: O(n) Time per lookup after that: O(1)
# Space: O(max_value
```

The character version — hashing letters 'a' to 'z'
Same idea, but the index trick uses ord(char) - ord('a') to map each lowercase letter to an index from 0 to 25

```python
def basic_hashing_chars(s):
  hash_arr = [0] * 26 # one slot per lowercase letter
  for ch in s:
    index = ord(ch) - ord('a') # 'a'->0, 'b'->1, ..., 'z'->25
    hash_arr[index] += 1

  return hash_arr
# Example usage:
s = "abcaab"
counts = basic_hashing_chars(s)
print(counts[ord('a') - ord('a')]) # count of 'a' -> 3
# Time: O(n) Space: O(26) = O(1)
```

Counting Frequencies of Array Elements
Problem: Given an array, count how many times each distinct element appears — the general-purpose version of section 3.2, not
restricted to a small known range

Approach A — Array-based hashing (when values are small non-negative integers, e.g. 1 to 1000)


```python
def count_frequencies_array(arr):
  max_val = max(arr)
  freq = [0] * (max_val + 1)
  for num in arr:
    freq[num] += 1
  return freq
# Time: O(n + max_val) Space: O(max_val)
arr=[1,2,2,3,4,4]
print(count_frequencies_array(arr))
```

```python
from collections import Counter
def count_frequencies_dict(arr):
  freq = {}
  for num in arr:
    freq[num] = freq.get(num, 0) + 1
  return freq
  # Or, using Python's built-in Counter (does exactly this internally):
  def count_frequencies_counter(arr):
    return Counter(arr)
# Example:
arr = [10, 20, 10, 30, 20, 10]
print(count_frequencies_dict(arr)) # {10: 3, 20: 2, 30: 1}
# Time: O(n) Space: O(number of distinct elements)
```

**3.4 Highest Occurring Element in an Array
Problem: Find the element that appears most frequently in the array. If there's a tie, any one of the tied elements is usually acceptable
(check the exact problem statement for tie-breaking rules).
Example: [1, 3, 2, 3, 4, 3, 2] → 3 (appears 3 times, more than anything else)**

Brute Force
For each element, count its occurrences by scanning the whole array again — O(n) work per element, O(n) elements to check → O(n²) total.

```python
def high_occ(arr):
  h_num,h_count=arr[0],0
  for num in arr:
    count = 0
    for x in arr:
      if x ==num:
        count+=1
    if count>h_count:
      h_num,h_count=num,count
  return h_num
arr=[2,3,4,5,3,4,5,2,3,2,3]
print(high_occ(arr))
# Time: O(n^2) Space: O(1)
```

Optimal — Build a frequency hashmap first, then scan it once

first build the frequency count (O(n)), then do a single linear scan over the counts to find the
maximum (O(n) again, but over the distinct elements, not squared).

```python
def highest_occurring(arr):
  freq = {}
  for num in arr:
    freq[num] = freq.get(num, 0) + 1
    best_val, best_count = None, 0
  for num, count in freq.items():
    if count > best_count:
      best_val, best_count = num, count
  return best_val
# Time: O(n) Space: O(n) for the frequency map
# One-liner using Counter's built-in most_common:
from collections import Counter
def highest_occurring_v2(arr):
  return Counter(arr).most_common(1)[0][0]
arr=[2,3,4,5,3,4,5,2,3,2,3]
print(highest_occurring(arr))
print(highest_occurring_v2(arr))
# Time: O(n) Space: O(n)
```
